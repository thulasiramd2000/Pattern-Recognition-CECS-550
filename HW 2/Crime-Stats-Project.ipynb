```python
X = df.drop('VictimCount', axis=1).values
Y = df['VictimCount'].values
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=52)

class CDS(Dataset):
    def __init__(self, Input, Output):
        self.Input = Input
        self.Output = Output

    def __len__(self):
        return len(self.Output)

    def __getitem__(self, idx):
        sample = {"Input": torch.tensor(self.Input[idx], dtype=torch.float32),
                  "Output": torch.tensor(self.Output[idx], dtype=torch.float32)}
        return sample
    
batch_size = 32

train_dataset = CDS(X_train, Y_train)
test_dataset = CDS(X_test, Y_test)

train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=batch_size, shuffle=False)   

for i, batch in enumerate(train_loader):
    print(f"Batch {i+1}:")
    print("Input shape:", batch['Input'].shape)
    print("Output shape:", batch['Output'].shape)
    if i == 1:  
        break
```

## Task 3
In this task you will try to predict number of crime victims as a **real number**. Therefore the machine learning problem is a **regression** problem. 

* Define the proper loss function for this task
* what should the size of the predicted output be?
* explain your choice of architecture, including how many layers you will be using
* define an optimizer for training this model, choose a proper learning rate 
* write a training loop that obtains a batch out of the  training data and calculates the forward and backward passes over the neural network. Call the optimizer to update the weights of the neural network.
* write a for loop that continues the training over a number of epochs. At the end of each epoch, calculate the ``MSE`` error on the test data and print it.
* is your model training well? Adjust the learning rate, hidden size of the network, and try different activation functions and number of layers to achieve the best accuracy and report it. 

```python
# For a regression problem which predicts a number of crime victims as a real number is MEAN SQUARED ERROR(MSE)
# MSE is the measure of average between actual value and predicted value
# MSE= 1/n ∑ n to i-1 (y1-y2)^2
#n is the total number of samples
#y1 is the actual value
#y2 is the predicted value

# A batch of 32 is defined in the dataloader, the output should be in the shape of ([32])

# Architecture
#Model = Feed Forward Neural Network
# Input layer=1, Hidden layers=2, Output Layer=1

class CrimePrediction(nn.Module):
    def __init__(self, input_size, hidden_size1, hidden_size2, output_size):
        super(CrimePrediction, self).__init__()
        
        self.fc1 = nn.Linear(input_size, hidden_size1)  
        self.fc2 = nn.Linear(hidden_size1, hidden_size2)  
        self.fc3 = nn.Linear(hidden_size2, output_size)
        self.relu = nn.ReLU()
        
    def forward(self, x):
        x = self.relu(self.fc1(x))
        x = self.relu(self.fc2(x))
        return self.fc3(x)

input_size = 6
hidden_size1 = 256
hidden_size2 = 128
output_size = 1

model = CrimePrediction(input_size, hidden_size1, hidden_size2, output_size)

learning_rate = 1e-4
optimizer = optim.Adam(model.parameters(), lr=learning_rate)

Lfunction = nn.MSELoss()

num_epochs = 18

for epoch in range(num_epochs):
    model.train()  
    total_train_loss = 0


    for i, batch in enumerate(train_loader):
        inputs = batch['Input']
        targets = batch['Output']
        
        optimizer.zero_grad()
        outputs = model(inputs)
        loss = Lfunction(outputs, targets)
        loss.backward()
        optimizer.step()
        total_train_loss += loss.item()

    average_train_loss = total_train_loss / len(train_loader)
    
    
    model.eval()  
    total_test_loss = 0
    with torch.no_grad():
        for batch in test_loader:
            inputs = batch['Input']
            targets = batch['Output']
            outputs = model(inputs)
            loss = Lfunction(outputs, targets)
            total_test_loss += loss.item()
    
    average_test_loss = total_test_loss / len(test_loader)

    print(f"Epoch [{epoch + 1}/{num_epochs}], Training Loss: {average_train_loss:.4f}, Test Loss: {average_test_loss:.4f}")
    
#REPORT
#Adjusted the learning rate from  0.001 to 0.0001
# Modified the hidden layers as 256 and 128
# used RELU

# Achieved :- Over the epochs, the loss typically declines from a relatively high starting point. 
#There is a general downward trend, with some variation in the middle. Positively, the final loss is significantly less than 
#the initial loss.

        
```

```python
class CrimeVictimClassification(nn.Module):
    def __init__(self, input_size, hidden_size1, hidden_size2, num_classes):
        super(CrimeVictimClassification, self).__init__()
        
        self.fc1 = nn.Linear(input_size, hidden_size1)  
        self.fc2 = nn.Linear(hidden_size1, hidden_size2)  
        self.fc3 = nn.Linear(hidden_size2, num_classes)  
        self.relu = nn.ReLU()
        
    def forward(self, x):
        x = self.relu(self.fc1(x))
        x = self.relu(self.fc2(x))
        x = self.fc3(x)  
        return x

input_size = 6
hidden_size1 = 256
hidden_size2 = 128
num_classes = 5
num_classes = len(set(Y_train))  
model = CrimeVictimClassification(input_size, hidden_size1, hidden_size2, num_classes)    

learning_rate = 1e-4
optimizer = optim.Adam(model.parameters(), lr=learning_rate)


loss_function = nn.CrossEntropyLoss()

num_epochs = 18
for epoch in range(num_epochs):
    total_loss = 0
    for i, batch in enumerate(train_loader):
        inputs = batch['Input']
        targets = batch['Output'].long()  
        optimizer.zero_grad()

        outputs = model(inputs)
        loss = loss_function(outputs, targets)
        loss.backward()
        optimizer.step()
        total_loss += loss.item()

    average_loss = total_loss / len(train_loader)
    print(f"Epoch [{epoch + 1}/{num_epochs}], Loss: {average_loss:.4f}")    
```

```python
def evaluate_model(model, loader, loss_function):
    model.eval()  
    total_loss = 0
    correct_predictions = 0
    total_samples = 0
    
    with torch.no_grad():
        for batch in loader:
            inputs = batch['Input']
            targets = batch['Output'].long()
            outputs = model(inputs)
            
            # Get the class label with highest probability as prediction
            _, predicted = outputs.max(1)
            
            total_loss += loss_function(outputs, targets).item()
            correct_predictions += (predicted == targets).sum().item()
            total_samples += targets.size(0)
    
    accuracy = 100 * correct_predictions / total_samples
    average_loss = total_loss / len(loader)
    return average_loss, accuracy

for epoch in range(num_epochs):
    total_loss = 0
    for i, batch in enumerate(train_loader):
        inputs = batch['Input']
        targets = batch['Output'].long()  
        optimizer.zero_grad()

        outputs = model(inputs)
        loss = loss_function(outputs, targets)
        loss.backward()
        optimizer.step()
        total_loss += loss.item()

    average_train_loss = total_loss / len(train_loader)

    average_test_loss, test_accuracy = evaluate_model(model, test_loader, loss_function)
    
    print(f"Epoch [{epoch + 1}/{num_epochs}], Training Loss: {average_train_loss:.4f}, Test Loss: {average_test_loss:.4f}, Test Accuracy: {test_accuracy:.2f}%")

    
    #Changed learning rate from 0.1 to 1e-4 and layers to 256 and 128 to achieve a highest accuracy rate of 54.55
```

```python
# From both the tasks of regression and classification model, for a typically small dataset regression might be more appropriate, 
#if there's a wide range and many outliers, classification into bins might be more stable.
#The number of victims as a continuous variable is directly predicted by regression, which can yield an accurate estimate. 
#Predicted values can yield an exact count that may be required for in-depth analysis or resource allocation.
#It may be simpler to train and comprehend the classification model if victims are grouped into buckets.

# Regression has less training loss and more accuracy than classification 
```

```python
def misclassification(model, loader):
    model.eval()
    misclassified_examples = []
    
    with torch.no_grad():
        for batch in loader:
            inputs = batch['Input']
            targets = batch['Output'].long()
            outputs = model(inputs)
            probabilities = nn.functional.softmax(outputs, dim=1)
            _, predicted = outputs.max(1)
            
            
            for i in range(len(targets)):
                if predicted[i] != targets[i]:
                    misclassified_examples.append((inputs[i], probabilities[i], targets[i]))
                    break

    return misclassified_examples[0] if misclassified_examples else None

misclassified_example = misclassification(model, test_loader)

if misclassified_example:
    inputs, probabilities, target = misclassified_example

    plt.bar(np.arange(num_classes), probabilities.numpy())
    plt.xlabel('Class Label')
    plt.ylabel('Probability')
    plt.title(f'Misclassified Example (True Label: {target.item()})')
    plt.xticks(np.arange(num_classes))
    plt.show()
else:
    print("No misclassified example found!")
```
